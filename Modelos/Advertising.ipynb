```python
import pandas as pd
import seaborn as sns
from seaborn import objects as so

import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# Colores corporativos Dataloop
PRIMARY = "#A2D738"
COMPLEMENTARY = "#FA1B74"
cmap = LinearSegmentedColormap.from_list("custom_palette",["white", PRIMARY, COMPLEMENTARY])
```

```python
df = pd.read_csv('Data/Advertising.csv')
df.head()   
```

### Visualización

Mapa de correlación:

```python
sns.heatmap(df.corr(), annot=True, cmap=cmap)
```

```python
(so.Plot(df['sales']).pair(y=[df['TV'], df['radio'], df['newspaper']]).add(so.Dots()))
```

```python
# Create a visualization
sns.relplot(
    data=df,
    x="TV", y="", col="time",
    hue="sales", style="smoker", size="size",
)
```
